```python
work_dir = '/content/drive/MyDrive/0_Lymphoma_PPFE_vs_Cryo/'          #you might need to change it in your case, look at your folder structure
img_dir = work_dir + 'images_1_CROPPED/'
mask_dir = work_dir + 'images_1_MASKS/'
```

```python
new_img_dir = work_dir + '0_CODES/alfa/small_images/images/'
new_mask_dir = work_dir + '0_CODES/alfa/small_images/masks/'
```

```python
images = sorted(glob(img_dir+"*_SELECTED_MODEL*"))                                #select only images containing SELECTED keyword in it
labels = sorted(glob(mask_dir+"*tif"))
labels.pop(9)
```

```python
ROW = 256
COL = 256
for k in range(len(labels)):
  img = io.imread(images[k])
  mask = io.imread(labels[k])
  mask[mask == 1] = 0
  mask = 255 - mask
  row = img.shape[0] // ROW
  col = img.shape[1] // COL
  for i in range(row):
    for j in range(col):
      temp_img_name = images[k].replace(img_dir, "").replace(".tif", "")
      temp_mask_name = labels[k].replace(mask_dir, "").replace(".tif", "")
      temp_img = img[i*ROW:(i+1)*ROW, j*COL:(j+1)*COL, :]
      temp_mask = mask[i*ROW:(i+1)*ROW, j*COL:(j+1)*COL]
      temp_img_1 = np.flip(temp_img)
      temp_mask_1 = np.flip(temp_mask)
      temp_img_2 = np.flip(temp_img, axis = 1)
      temp_mask_2 = np.flip(temp_mask, axis = 1)
      io.imsave(new_img_dir + temp_img_name + "_" + str(i)+str(j)+".tif", temp_img)
      io.imsave(new_mask_dir + temp_mask_name + "_" + str(i)+str(j)+".tif", temp_mask)
      io.imsave(new_img_dir + temp_img_name + "_" + str(i)+str(j)+"_flip_1"+".tif", temp_img_1)
      io.imsave(new_mask_dir + temp_mask_name + "_" + str(i)+str(j)+"_flip_1"+".tif", temp_mask_1)
      io.imsave(new_img_dir + temp_img_name + "_" + str(i)+str(j)+"_flip_2"+".tif", temp_img_2)
      io.imsave(new_mask_dir + temp_mask_name + "_" + str(i)+str(j)+"_flip_2"+".tif", temp_mask_2)
  # break
```
